```python
ok = 0
dataNames = []
error = []
tooMuchColumns = 0
tooManyClass = 0
for j in data:
    try:
        df1 = j.get_data(dataset_format ='dataframe')
        df = df1[0]
        if j.name in dataNames:
            continue
        dataNames.append(j.name)
        for i in df.columns:
            # print(df[i].dtype)
            if (df[i].dtype == object or df[i].dtype == 'category' or df[i].dtype == 'string'):
                if(len(df[i].unique()) <= 2):
                    df[i] = (df[i] == df[i].iloc[0])*1
        className = df.columns[len(df.columns)-1]
        classValues = df[className]
        
        if (len(df[className].unique()) != 2):
            tooManyClass +=1
            continue        
        
        df = df.drop(className, axis=1)
        df = pd.get_dummies(df)

        numberColumns = len(df.columns)
        if (numberColumns > 200):
            tooMuchColumns+=1
            continue

        df.columns = list([f'X_{i}' for i in range(numberColumns)])

        df['class'] = classValues
        # print(f'NoiseSelection/Datasets/{str(j.id)}_{j.name}.csv.zip')
        df.to_csv(f'Datasets/{str(j.id)}_{j.name}.csv.zip', index=False)
        ok+=1
        
    except:
        error.append(j.name)
```

```python

def getFolderFilesList(path):
    return [filename for filename in listdir(path) if isfile(join(path, filename))]

def addNoiseToDf(df, atribute, percent):
    df_copy = df.copy()
    labels = df[atribute].unique()
    sizeDf = len(df.index)
    quantityOfNoise = int((sizeDf*(percent)/100))
    listOfChangedIndex = random.sample(range(0, sizeDf - 1), quantityOfNoise)
    for index in listOfChangedIndex:
        newLabel = labels[1] if df[atribute][index] == labels[0] else labels[0]
        df_copy[atribute][index] = newLabel
    return df_copy
```

```python
listOfFiles = getFolderFilesList('Datasets')
for file in listOfFiles:
    df = pd.read_csv(f'Datasets/{file}')
    for i in range(10):
        df5 = addNoiseToDf(df, 'class', 5)
        df10 = addNoiseToDf(df, 'class', 10)
        df20 = addNoiseToDf(df, 'class', 20)
        # df40 = addNoiseToDf(df, 'class', 40)
        dfName = file.split('.')[0]
        df5.to_csv(f'NoisyDatasets/{dfName}_5_{str(i)}.csv.gz', index=False)
        df10.to_csv(f'NoisyDatasets/{dfName}_10_{str(i)}.csv.gz', index=False)
        df20.to_csv(f'NoisyDatasets/{dfName}_20_{str(i)}.csv.gz', index=False)
        # df40.to_csv(f'NoisyDatasets/{dfName}_40_{str(i)}.csv.gz', index=False)
```

```python
# extracts meta-features from clear datasets (without noise)
datasetsFilesNames = getFolderFilesList("NoisyDatasets")
listOfMfeGroups = ["general", "statistical", "info-theory", "model-based", "landmarking"]
data = pd.DataFrame()
count = 0
for dataset in datasetsFilesNames:
    df =  pd.read_csv("NoisyDatasets/" + dataset)
    mfe = MFE(groups=listOfMfeGroups, num_cv_folds= 5)
    X, y = df.drop('class', axis=1), df['class']
    mfe.fit(X.values, y.values)
    ft = mfe.extract()
    df_MFE = pd.DataFrame(columns=ft[0], data = [ft[1]])
    df_MFE["Dataset"] = dataset
    data = data.append(df_MFE)
    clear_output(wait=True)
    count +=1
    print ("exec " + str(count) + " / " + str(len(datasetsFilesNames)))

# data.dropna(axis=1)
data.to_csv("meta_features.csv",index=False)
data.dropna(axis=1).to_csv("meta_features_dropna.csv",index=False)
```
